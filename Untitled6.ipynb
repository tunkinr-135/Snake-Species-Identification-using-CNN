```python
#@title function evaluate_model
res = {}
def evaluate_model(model,y_test, y_pre):
    """
    to evaluate the model

    """
    results = {}

    accuracy = accuracy_score(y_test, y_pre)
    results['accuracy'] = accuracy

    f1 = f1_score(y_test, y_pre, average='weighted')
    results['f1_score'] = f1

    recall = recall_score(y_test, y_pre, average='weighted')
    results['recall'] = recall

    precision = precision_score(y_test, y_pre, average='weighted')
    results['precision'] = precision
    res[model] = results
    return results
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score, roc_curve
from sklearn.utils import class_weight
from imblearn.over_sampling import SMOTE      # <<< CHANGED HERE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, LeakyReLU, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
import tensorflow as tf
```

```python
 def focal_loss(gamma=2., alpha=0.25):
    def fl(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        y_pred = tf.clip_by_value(y_pred, 1e-7, 1 - 1e-7)
        return tf.reduce_mean(
            -alpha * (1 - y_pred)**gamma * y_true * tf.math.log(y_pred)
            - (1 - alpha) * y_pred**gamma * (1 - y_true) * tf.math.log(1 - y_pred)
        )
    return fl
```

```python
target_col = 'Diagnosis'
X = df1.drop(target_col, axis=1)
y = df1[target_col]
```

```python
# Encode target
if y.dtype == 'object':
    y = LabelEncoder().fit_transform(y)
```

```python
# One-hot encoding
X = pd.get_dummies(X, drop_first=True)
```

```python
# FAST RESAMPLING
smote = SMOTE(random_state=42)  # <<< SMOTE ONLY
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Train-test split (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)
```

```python
# Scale
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Class weights
classes = np.unique(y_train)
weights = class_weight.compute_class_weight('balanced', classes=classes, y=y_train)
class_weights = {int(c): float(w) for c, w in zip(classes, weights)}
```

```python
# Callbacks
early_stop = EarlyStopping(patience=12, restore_best_weights=True, verbose=1)
reduce_lr = ReduceLROnPlateau(patience=6, factor=0.5, min_lr=1e-6, verbose=1)
```

```python
# MODEL
model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(512, kernel_initializer='he_normal', kernel_regularizer=l2(0.001)),
    BatchNormalization(), LeakyReLU(), Dropout(0.4),
    Dense(256, kernel_initializer='he_normal', kernel_regularizer=l2(0.001)),
    BatchNormalization(), LeakyReLU(), Dropout(0.4),
    Dense(256, kernel_initializer='he_normal'),
    BatchNormalization(), LeakyReLU(), Dropout(0.3),
    Dense(128, kernel_initializer='he_normal'),
    BatchNormalization(), LeakyReLU(), Dropout(0.3),
    Dense(128, kernel_initializer='he_normal'),
    BatchNormalization(), LeakyReLU(), Dropout(0.3),
    Dense(64, kernel_initializer='he_normal'),
    BatchNormalization(), LeakyReLU(), Dropout(0.25),
    Dense(64, kernel_initializer='he_normal'),
    BatchNormalization(), LeakyReLU(), Dropout(0.25),
    Dense(32, kernel_initializer='he_normal'),
    BatchNormalization(), LeakyReLU(), Dropout(0.2),
    Dense(16, kernel_initializer='he_normal'),
    BatchNormalization(), LeakyReLU(), Dropout(0.2),
    Dense(8, activation='relu'), Dropout(0.1),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss=focal_loss(2,0.25), metrics=['accuracy'])
```

```python
# TRAIN
history = model.fit(
    X_train, y_train,
    validation_split=0.15,
    epochs=150,
    batch_size=128,
    class_weight=class_weights,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)
```

```python
# Predict
y_prob = model.predict(X_test).ravel()
```

```python
# Optimal threshold
fpr, tpr, thr = roc_curve(y_test, y_prob)
opt_idx = np.argmax(tpr - fpr)
opt_thr = thr[opt_idx]
```

```python
# Apply threshold
y_pred = (y_prob > opt_thr).astype(int)
```

```python
evaluate_model('ANN', y_test, y_pred)
print(classification_report(y_test, y_pred))
```

```python
# ============================================================
# 1) FIXED XGBoost training with old-version compatibility
# ============================================================

from xgboost import XGBClassifier
from sklearn.metrics import (
    roc_auc_score, roc_curve, accuracy_score,
    classification_report, confusion_matrix,
    precision_score, recall_score, f1_score
)
import numpy as np


# === Build model (NO early_stopping_rounds here) ===
final_xgb = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42,
    n_estimators=150,
    max_depth=3,
    learning_rate=0.06,
    subsample=1.0,
    colsample_bytree=0.8,
    gamma=0,
    reg_alpha=0.5,
    reg_lambda=1,
    max_delta_step=1
)

# === Set early stopping via set_params (works for ALL versions) ===
final_xgb.set_params(early_stopping_rounds=30)

print(" Training Lightened XGBoost Model (compatible early stopping)...")

final_xgb.fit(
    X_tr, y_tr,
    eval_set=[(X_val, y_val)],
    verbose=False
)


# ============================================================
# 2) Get ANN + XGB probabilities
# ============================================================

p_ann_val  = model.predict(X_val).ravel()
p_ann_test = model.predict(X_test).ravel()

p_xgb_val  = final_xgb.predict_proba(X_val)[:,1]
p_xgb_test = final_xgb.predict_proba(X_test)[:,1]


# ============================================================
# 3) Compute validation AUCs  ensemble weights
# ============================================================

auc_ann_val = roc_auc_score(y_val, p_ann_val)
auc_xgb_val = roc_auc_score(y_val, p_xgb_val)

print(f"Validation AUCs  ANN: {auc_ann_val:.4f}, XGB: {auc_xgb_val:.4f}")

weights = np.array([auc_ann_val, auc_xgb_val])
weights = weights / weights.sum()

print("Ensemble weights (ANN, XGB):", np.round(weights, 3))


# ============================================================
# 4) Tune threshold on validation set
# ============================================================

ensemble_val_prob = weights[0]*p_ann_val + weights[1]*p_xgb_val

fpr, tpr, thr = roc_curve(y_val, ensemble_val_prob)
opt_idx = np.argmax(tpr - fpr)
opt_thr = thr[opt_idx]

print(f"Optimal validation threshold: {opt_thr:.4f}")


# ============================================================
# 5) Apply ensemble to test set
# ============================================================

ensemble_test_prob = weights[0]*p_ann_test + weights[1]*p_xgb_test
ensemble_test_pred = (ensemble_test_prob >= opt_thr).astype(int)


# ============================================================
# 6) Evaluate on test set
# ============================================================

print("\n=== ENSEMBLE RESULTS (TEST SET) ===")
print("AUC:", roc_auc_score(y_test, ensemble_test_prob))
print("Accuracy:", accuracy_score(y_test, ensemble_test_pred))

print("\nConfusion Matrix:\n", confusion_matrix(y_test, ensemble_test_pred))
print("\nClassification Report:\n", classification_report(y_test, ensemble_test_pred))

print("Precision:", precision_score(y_test, ensemble_test_pred))
print("Recall   :", recall_score(y_test, ensemble_test_pred))
print("F1 Score :", f1_score(y_test, ensemble_test_pred))
```

```python
# Complete ensemble pipeline (RF1 + sklearn-MLP + XGBoost + ANN) with proper calibration
import warnings
warnings.filterwarnings('ignore')

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    roc_auc_score, roc_curve, accuracy_score, classification_report, confusion_matrix
)
from xgboost import XGBClassifier

# --------------- Ensure train/val split exists ---------------
try:
    X_tr, X_val, y_tr, y_val  # check variables
    _ = X_tr  # if defined, nothing happens
except Exception:
    # create a validation split from your existing X_train, y_train
    print("Creating X_tr / X_val split from X_train...")
    X_tr, X_val, y_tr, y_val = train_test_split(
        X_train, y_train, test_size=0.20, stratify=y_train, random_state=42
    )

# --------------- Train RandomForest (rf1_model) if missing ---------------
if 'rf1_model' not in globals():
    print("Training RandomForest (rf1_model)...")
    rf1_model = RandomForestClassifier(
        n_estimators=300,
        class_weight='balanced',
        random_state=42,
        n_jobs=-1
    )
    rf1_model.fit(X_tr, y_tr)

# --------------- Train sklearn MLP (sk_mlp) if missing ---------------
if 'sk_mlp' not in globals():
    print("Training sklearn MLP (sk_mlp)...")
    sk_mlp = MLPClassifier(
        hidden_layer_sizes=(256,128),
        max_iter=400,
        random_state=42,
        early_stopping=True,
        n_iter_no_change=20,
        batch_size=128
    )
    sk_mlp.fit(X_tr, y_tr)

# --------------- Train XGBoost (final_xgb) if missing ---------------
if 'final_xgb' not in globals():
    print("Training XGBoost (final_xgb)...")
    final_xgb = XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42,
        n_estimators=150,
        max_depth=3,
        learning_rate=0.06,
        subsample=1.0,
        colsample_bytree=0.8,
        gamma=0,
        reg_alpha=0.5,
        reg_lambda=1,
        max_delta_step=1
    )
    # set early stopping via set_params for compatibility across xgboost versions
    final_xgb.set_params(early_stopping_rounds=30)
    final_xgb.fit(X_tr, y_tr, eval_set=[(X_val, y_val)], verbose=False)

# --------------- Calibration ---------------
print("Calibrating RF1 and sklearn-MLP (cv=3 isotonic)...")
cal_rf1 = CalibratedClassifierCV(rf1_model, cv=3, method='isotonic')
cal_mlp = CalibratedClassifierCV(sk_mlp, cv=3, method='isotonic')

cal_rf1.fit(X_tr, y_tr)
cal_mlp.fit(X_tr, y_tr)

# Calibrate XGBoost using the prefit approach on X_val (no retraining)
print("Calibrating XGBoost using cv='prefit' on X_val (no retrain)...")
cal_xgb = CalibratedClassifierCV(final_xgb, cv='prefit', method='isotonic')
cal_xgb.fit(X_val, y_val)

# Use calibrated classifiers going forward
rf1_clf = cal_rf1
sk_mlp_clf = cal_mlp
xgb_clf = cal_xgb

# --------------- Probabilities on val & test ---------------
print("Generating probabilities on validation and test sets...")
p_ann_val  = model.predict(X_val).ravel()
p_ann_test = model.predict(X_test).ravel()

p_rf1_val  = rf1_clf.predict_proba(X_val)[:,1]
p_rf1_test = rf1_clf.predict_proba(X_test)[:,1]

p_xgb_val  = xgb_clf.predict_proba(X_val)[:,1]
p_xgb_test = xgb_clf.predict_proba(X_test)[:,1]

p_mlp_val  = sk_mlp_clf.predict_proba(X_val)[:,1]
p_mlp_test = sk_mlp_clf.predict_proba(X_test)[:,1]

print("Validation AUCs:")
print("ANN :", round(roc_auc_score(y_val, p_ann_val),4))
print("RF1 :", round(roc_auc_score(y_val, p_rf1_val),4))
print("XGB :", round(roc_auc_score(y_val, p_xgb_val),4))
print("MLP :", round(roc_auc_score(y_val, p_mlp_val),4))

# --------------- Blending (meta-learner) ---------------
print("\nTraining LogisticRegression meta-learner on validation predictions...")
X_meta_val = np.vstack([p_ann_val, p_rf1_val, p_xgb_val, p_mlp_val]).T
X_meta_test = np.vstack([p_ann_test, p_rf1_test, p_xgb_test, p_mlp_test]).T

meta = LogisticRegression(max_iter=4000)
meta.fit(X_meta_val, y_val)

meta_prob_test = meta.predict_proba(X_meta_test)[:,1]
meta_prob_val  = meta.predict_proba(X_meta_val)[:,1]

# tune threshold on val meta predictions (maximize TPR - FPR)
fpr, tpr, thr = roc_curve(y_val, meta_prob_val)
opt_thr_meta = thr[np.argmax(tpr - fpr)]

meta_pred_test = (meta_prob_test >= opt_thr_meta).astype(int)

print("\n--- Blending (Logistic meta) results on TEST ---")
print("Used validation-optimal threshold:", round(opt_thr_meta,4))
print("AUC:", round(roc_auc_score(y_test, meta_prob_test),4))
print("Accuracy:", round(accuracy_score(y_test, meta_pred_test),4))
print(classification_report(y_test, meta_pred_test))
print("Confusion Matrix:\n", confusion_matrix(y_test, meta_pred_test))

# --------------- Grid-search weight for ANN vs others (fast) ---------------
print("\nRunning ANN-weight grid search (w from 0 to 1 step 0.05) ...")
p_others_val  = (p_rf1_val + p_xgb_val + p_mlp_val) / 3.0
p_others_test = (p_rf1_test + p_xgb_test + p_mlp_test) / 3.0

best = {'w':None, 'val_acc':-1, 'test_acc':-1, 'test_auc':-1}
ws = np.linspace(0,1,21)
for w in ws:
    prob_val = w * p_ann_val + (1-w) * p_others_val
    fpr_v, tpr_v, thr_v = roc_curve(y_val, prob_val)
    opt_thr_v = thr_v[np.argmax(tpr_v - fpr_v)]
    pred_val = (prob_val >= opt_thr_v).astype(int)
    val_acc = accuracy_score(y_val, pred_val)
    if val_acc > best['val_acc']:
        prob_test = w * p_ann_test + (1-w) * p_others_test
        pred_test = (prob_test >= opt_thr_v).astype(int)
        best.update({
            'w': float(w),
            'val_acc': val_acc,
            'test_acc': accuracy_score(y_test, pred_test),
            'test_auc': roc_auc_score(y_test, prob_test),
            'thr': float(opt_thr_v)
        })

print("Best ANN weight (grid):", best['w'])
print("Validation Acc (grid):", round(best['val_acc'],4))
print("Corresponding Test Acc:", round(best['test_acc'],4), " Test AUC:", round(best['test_auc'],4))
print("Threshold used from val:", round(best['thr'],4))

# --------------- Summary ---------------
stacking_acc = accuracy_score(y_test, meta_pred_test)
stacking_auc = roc_auc_score(y_test, meta_prob_test)
grid_acc = best['test_acc']
grid_auc = best['test_auc']

print("\n--- Summary ---")
print(f"Blending (meta) -> Test Acc: {stacking_acc:.4f}, AUC: {stacking_auc:.4f}")
print(f"Grid-weight   -> Test Acc: {grid_acc:.4f}, AUC: {grid_auc:.4f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, roc_auc_score, roc_curve

# ---- helper: find optimal threshold on val (maximize TPR - FPR) ----
def find_opt_thr(yv, pv):
    fpr, tpr, thr = roc_curve(yv, pv)
    idx = np.argmax(tpr - fpr)
    return thr[idx]

# ---- compute thresholds for each base model using validation set ----
thr_ann = find_opt_thr(y_val, p_ann_val)
thr_xgb = find_opt_thr(y_val, p_xgb_val)
thr_rf1 = find_opt_thr(y_val, p_rf1_val)
thr_mlp = find_opt_thr(y_val, p_mlp_val)
# meta used earlier opt_thr_meta (if exists) otherwise compute on meta_val
try:
    opt_thr_meta  # if variable exists from earlier cell
except NameError:
    opt_thr_meta = find_opt_thr(y_val, meta_prob_val)

# ---- create binary predictions on test using those thresholds ----
pred_ann = (p_ann_test >= thr_ann).astype(int)
pred_xgb = (p_xgb_test >= thr_xgb).astype(int)
pred_rf1 = (p_rf1_test >= thr_rf1).astype(int)
pred_mlp = (p_mlp_test >= thr_mlp).astype(int)
pred_meta = (meta_prob_test >= opt_thr_meta).astype(int)  # already available as meta_pred_test but safe

models = {
    'ANN'   : (pred_ann, p_ann_test),
    'XGBoost':(pred_xgb, p_xgb_test),
    'RF1'   : (pred_rf1, p_rf1_test),
    'MLP'   : (pred_mlp, p_mlp_test),
    'Ensemble (meta)': (pred_meta, meta_prob_test)
}

# ---- helper to compute per-class metrics ----
def class_metrics(y_true, y_pred):
    classes = np.unique(y_true)
    # assume binary {0,1} with 0=Benign,1=Malignant
    prec0 = precision_score(y_true, y_pred, pos_label=classes[0])
    rec0  = recall_score(y_true, y_pred, pos_label=classes[0])
    f10   = f1_score(y_true, y_pred, pos_label=classes[0])
    prec1 = precision_score(y_true, y_pred, pos_label=classes[1])
    rec1  = recall_score(y_true, y_pred, pos_label=classes[1])
    f11   = f1_score(y_true, y_pred, pos_label=classes[1])
    acc   = accuracy_score(y_true, y_pred)
    return {
        'benign': {'precision': prec0, 'recall': rec0, 'f1': f10},
        'malignant': {'precision': prec1, 'recall': rec1, 'f1': f11},
        'accuracy': acc
    }

# ---- Plotting ----
metrics = ['Precision','Recall','F1-Score','Accuracy']
n_models = len(models)
cols = 3
rows = int(np.ceil(n_models/cols))

fig, axes = plt.subplots(rows, cols, figsize=(cols*5.2, rows*4.2))
axes = axes.flatten()

colors = {'benign':'#66c2ff', 'malignant':'#ff9999'}

for ax, (mname, (mpred, mprob)) in zip(axes, models.items()):
    m = class_metrics(y_test, mpred)
    benign_vals = [m['benign']['precision'], m['benign']['recall'], m['benign']['f1'], m['accuracy']]
    malign_vals = [m['malignant']['precision'], m['malignant']['recall'], m['malignant']['f1'], m['accuracy']]
    x = np.arange(len(metrics))
    width = 0.35
    bars0 = ax.bar(x - width/2, benign_vals, width, label='Benign', color=colors['benign'])
    bars1 = ax.bar(x + width/2, malign_vals, width, label='Malignant', color=colors['malignant'])
    # annotate
    for b in list(bars0)+list(bars1):
        h = b.get_height()
        ax.annotate(f"{h:.2f}", xy=(b.get_x()+b.get_width()/2, h), xytext=(0,6),
                    textcoords='offset points', ha='center', va='bottom', fontsize=9)
    ax.set_xticks(x)
    ax.set_xticklabels(metrics, fontsize=10)
    ax.set_ylim(0,1.05)
    ax.set_title(f"{mname}\nAUC: {roc_auc_score(y_test, mprob):.3f}", fontsize=11)
    ax.grid(axis='y', linestyle='--', alpha=0.4)
    ax.legend(fontsize=9)

# hide unused subplots
for i in range(len(models), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.savefig('ensemble_model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```
